 ### Sales Analysis: Yearly Analysis, Product Filters, and Sales Categorization

**Description:**

This SQL query performs a sales analysis on the superstore dataset, focusing on a specific date range and multiple product names. It includes the following key features:

**Date Filter:** The query filters the dataset to include only the orders placed in the year 2012. This allows for a specific analysis based on a particular year.

**Multiple Text Filters:

The query applies a text filter to the "Product Name" column, selecting specific products based on their names. Products containing the terms "bulb," "Lock," "print," "wire," "copier," "trays," or "trimmer" are included in the analysis.

**Addition of Sales Category Column:** 

The query adds a new column called "SalesCategory" that categorizes the sales based on their values. Sales above 250 are labeled as "Highest Sales," sales between 150 and 250 are labeled as "Medium Sales," and the rest are labeled as "Lowest Sales."

**Changing Ship Mode Text:** 

The query modifies the text in the "Ship Mode" column to replace specific ship mode names. "First Class" is changed to "First," "Same Day" is changed to "Same," "Standard Class" is changed to "Standard," and "Second Class" is changed to "Second." This transformation simplifies the ship mode names for better readability.

This query provides valuable insights into sales performance by applying filters on dates, product names, and ship modes, while also categorizing sales.


```python
df = execute_query("SELECT YEAR(OrderDate) AS YearNumber, FORMAT(OrderDate, 'MMMM') AS MonthName, Category, [Product Name], Country, FORMAT(Sales, 'N2') AS Sales, CASE WHEN Sales > 250 THEN 'Highest Sales' WHEN Sales BETWEEN 150 AND 250 THEN 'Medium Sales' ELSE 'Lowest Sales' END AS SalesCategory, CASE WHEN \"Ship Mode\" = 'First Class' THEN 'First' WHEN \"Ship Mode\" = 'Same Day' THEN 'Same' WHEN \"Ship Mode\" = 'Standard Class' THEN 'Standard' WHEN \"Ship Mode\" = 'Second Class' THEN 'Second' END AS Ship_Mode FROM dbo.superstore_dataset WHERE (YEAR(OrderDate) = 2012) AND ([Product Name] LIKE '%bulb%' OR [Product Name] LIKE '%Lock%' OR [Product Name] LIKE '%print%' OR [Product Name] LIKE '%wire%' OR [Product Name] LIKE '%copier%' OR [Product Name] LIKE '%trays%' OR [Product Name] LIKE '%trimmer%') AND COUNTRY = 'India' ORDER BY CAST(Sales AS DECIMAL(10,2)) DESC;")
df
```

```python
# Convert 'Sales' column to numeric
df['Sales'] = df['Sales'].str.replace(',', '').astype(float)

# Group the data by Category and SalesCategory and calculate the total sales
grouped_df = df.groupby(['Category', 'SalesCategory']).agg({'Sales': 'sum'}).reset_index()

# Pivot the data to create a stacked bar chart
pivot_df = grouped_df.pivot(index='Category', columns='SalesCategory', values='Sales')

# Set the plot style
sns.set(style="whitegrid")

# Plot the stacked bar chart
plt.figure(figsize=(12, 6))
pivot_df.plot(kind='bar', stacked=True)
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Sales by Category and Sales Category')
plt.legend(title='Sales Category')
plt.tight_layout()
plt.show()
```
